```python
table = soup.find("table", id="t3-rajasthan-royals")

# Headers
header = [th.get_text(strip=True) for th in table.find("thead").find_all("th")]
header = header[:-1]          # Remove last empty heading

# Create DataFrame
df = pd.DataFrame(columns=header)

# Rows
rows = table.find("tbody").find_all("tr")

for tr in rows:
    row = [td.get_text(strip=True) for td in tr.find_all("td")]
    row = row[:-1]            # Remove last empty value
    df.loc[len(df)] = row

print(df)
```

```python
#Scratch data from all 14 tables
import pandas as pd

all_data = []

# Get all team tables
tables = soup.find_all("table", class_="sold-players")

for table in tables:

    # Team name from table id
    team = table.get("id").replace("t3-", "").replace("-", " ").title()

    rows = table.find("tbody").find_all("tr")

    for tr in rows:
        row = [td.get_text(strip=True) for td in tr.find_all("td")]

        # Remove last empty column
        row = row[:-1]

        # Add team name as first column
        row.insert(0, team)

        all_data.append(row)

# Create DataFrame
columns = [
    "Team",
    "Sr. No.",
    "Player",
    "Base Price",
    "Winning Bid",
    "Capped/Uncapped"
]

df = pd.DataFrame(all_data, columns=columns)

print(df)
```
